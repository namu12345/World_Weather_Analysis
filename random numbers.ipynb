```python
def latitudes(size):
    latitudes = []
    x = 0
    while x < (size):
        random_lat = rd.randint(-90, 90) + rd.random()
        latitudes.append(random_lat)
        x += 1
    return latitudes
# Call the function with 1500.
%timeit latitudes(1500)
```

```python
def latitudes125(size):
    latitudes1 = []
    for x in range(size):
        random_lat_1 = rd.randint(-90, 90) + rd.random()
        latitudes1.append(random_lat_1)
    
    print(latitudes1)
    
latitudes125(1500)
```
